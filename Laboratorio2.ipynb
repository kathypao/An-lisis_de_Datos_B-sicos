```python
 import pandas as pd

# cargar el Dataset
df = pd.read_csv('CSV Laboratorio 2 AD-B.csv')

# Mostrar las primeras filas del dataFrame

print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
 #Crear columna precio_total
df['Precio_total'] = df['Precio_unitario']* df['Cantidad_vendida']
print(df)
```

```python
# Filtrar productos de los que se hayan vendido mas de 40 unidades
df_filtrado = df[df['Cantidad_vendida'] > 40]
print(df_filtrado)
```

```python
# Filtrar los 10 precios más altos
df_ordenado = df.sort_values(by='Precio_unitario', ascending=False)
df_top_10 = df_ordenado.head(10)
print(df_top_10)
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Agrupar ventas por sucursal y mes, sumando las cantidades vendidas
df_resumen = df.groupby(["Sucursal", "Mes"])["Cantidad_vendida"].sum()

#Crear un gráfico de barras
df_resumen.plot(kind="bar", figsize=(10, 6 ),width=0.8, colormap="viridis")

# Personalizar grafico
plt.title("Ventas por sucursal y meses", fontsize=14)
plt.xlabel("Sucursales", fontsize=10)
plt.ylabel("Cantidad vendida", fontsize=9)
plt.xticks(rotation=0)
plt.legend(title="Meses", bbox_to_anchor=(1.05,1), loc="upper left")
plt.tight_layout()
plt.grid(axis="y", linestyle="-", alpha=0.7)

# Mostrar gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear gráfico de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(df["Precio_unitario"], df["Cantidad_vendida"], color='blue', alpha=0.8, edgecolor='black')

# Añadir etiquetas y título
plt.xlabel("Precio unitario", fontsize=9)
plt.ylabel("Cantidad vendida", fontsize=9)
plt.grid(alpha=0.3)
plt.title("Relación entre Precio Unitario y cantidad vendia", fontsize=10)


# Mostrar gráfico
plt.show()
```
